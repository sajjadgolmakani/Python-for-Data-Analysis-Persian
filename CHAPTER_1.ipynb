# 1.1 این کتاب درباره چیست؟

این کتاب به جنبه‌های عملی و فنیِ **دست‌کاری (manipulating)**، **پردازش (processing)**، **پاک‌سازی (cleaning)** و **تحلیل (crunching)** داده‌ها در زبان برنامه‌نویسی پایتون اختصاص دارد. هدف من ارائه‌ی راهنمایی جامع در مورد بخش‌های کلیدی زبان پایتون و اکوسیستم کتابخانه‌ها و ابزارهای داده‌محور آن است؛ ابزارهایی که به شما کمک می‌کنند تا به یک **تحلیلگر داده (data analyst)** مؤثر تبدیل شوید.

اگرچه «تحلیل داده» در عنوان کتاب آمده است، تمرکز اصلی بر روی **برنامه‌نویسی پایتون**، کتابخانه‌ها و ابزارهای آن است و نه بر روی روش‌شناسی تحلیل داده. این همان بخش از پایتون است که شما برای انجام تحلیل داده به آن نیاز دارید.

مدتی پس از انتشار اولیه‌ی این کتاب در سال ۲۰۱۲، اصطلاح **علم داده (data science)** به‌عنوان یک اصطلاح چترگونه برای توصیف همه‌چیز از آمار توصیفی ساده تا تحلیل‌های آماری پیشرفته و **یادگیری ماشین (machine learning)** رواج پیدا کرد. از آن زمان تاکنون، اکوسیستم متن‌باز پایتون برای تحلیل داده یا علم داده نیز به شکل چشمگیری رشد کرده است. اکنون منابع بسیاری وجود دارند که به‌طور خاص بر روی روش‌های پیشرفته‌تر تمرکز دارند. امید من این است که این کتاب، پایه‌ای محکم برای حرکت شما به‌سوی منابع تخصصی‌تر در حوزه‌های خاص فراهم آورد.

برخی ممکن است محتوای این کتاب را بیشتر به‌عنوان «دست‌کاری داده» (data manipulation) توصیف کنند تا «تحلیل داده». همچنین از واژه‌های **wrangling** یا **munging** نیز برای اشاره به همین مفهوم استفاده می‌شود.

## چه نوع داده‌هایی مد نظر است؟

وقتی من از «داده» صحبت می‌کنم، دقیقاً منظورم چیست؟ تمرکز اصلی بر روی **داده‌های ساخت‌یافته (structured data)** است؛ اصطلاحی عامدانه مبهم که طیف وسیعی از انواع رایج داده‌ها را شامل می‌شود، از جمله:

* داده‌های جدولی یا شبیه صفحه‌گسترده (spreadsheet) که در آن هر ستون می‌تواند نوع متفاوتی داشته باشد (رشته‌ای، عددی، تاریخی و غیره). این شامل بیشتر داده‌هایی است که معمولاً در پایگاه‌های داده رابطه‌ای یا فایل‌های متنی جداسازی‌شده با تب یا کاما ذخیره می‌شوند.
* آرایه‌های چندبعدی (matrices).
* چندین جدول داده که از طریق ستون‌های کلید با هم مرتبط هستند (چیزی شبیه کلیدهای اصلی یا خارجی در SQL).
* سری‌های زمانی با فواصل منظم یا نامنظم.

البته این فهرست کامل نیست. اگرچه همیشه آشکار نیست، اما درصد بالایی از مجموعه‌داده‌ها را می‌توان به شکلی ساخت‌یافته تبدیل کرد که برای تحلیل و مدل‌سازی مناسب‌تر باشد. اگر چنین نباشد، اغلب می‌توان **ویژگی‌هایی (features)** از داده‌ها استخراج کرد تا آن‌ها را به شکلی ساخت‌یافته درآورد.
برای مثال، مجموعه‌ای از مقالات خبری را می‌توان به یک جدول فراوانی واژگان تبدیل کرد که سپس می‌تواند برای **تحلیل احساسات (sentiment analysis)** به کار رود.

اکثر کاربران نرم‌افزارهای صفحه‌گسترده مانند Microsoft Excel، که شاید پرکاربردترین ابزار تحلیل داده در جهان باشد، با این نوع داده‌ها آشنایی دارند.

# 1.2 چرا پایتون برای تحلیل داده؟

برای بسیاری از افراد، زبان برنامه‌نویسی **پایتون (Python)** جذابیت زیادی دارد. از زمان انتشار نخستین نسخه در سال ۱۹۹۱، پایتون به یکی از محبوب‌ترین زبان‌های برنامه‌نویسی تفسیری تبدیل شده است؛ در کنار زبان‌هایی مانند Perl و Ruby. از حدود سال ۲۰۰۵ به بعد، پایتون و روبی به‌ویژه به دلیل فریم‌ورک‌های وب متعدد خود—مانند Ruby on Rails برای روبی و Django برای پایتون—برای ساخت وب‌سایت‌ها بسیار محبوب شدند.

چنین زبان‌هایی اغلب «زبان اسکریپتی (scripting languages)» نامیده می‌شوند، زیرا می‌توان با آن‌ها به‌سرعت برنامه‌های کوچک یا اسکریپت‌هایی برای خودکارسازی وظایف نوشت. من شخصاً با این اصطلاح موافق نیستم، زیرا این برداشت را القا می‌کند که این زبان‌ها برای توسعه‌ی نرم‌افزارهای جدی مناسب نیستند.

در میان زبان‌های تفسیری، به دلایل تاریخی و فرهنگی مختلف، پایتون یک جامعه‌ی فعال و گسترده در زمینه‌ی **محاسبات علمی (scientific computing)** و تحلیل داده شکل داده است. در ۲۰ سال گذشته، پایتون از یک زبان محاسباتی پیشرو و پرریسک به یکی از مهم‌ترین زبان‌ها برای **علم داده، یادگیری ماشین** و توسعه‌ی عمومی نرم‌افزار در دانشگاه‌ها و صنعت تبدیل شده است.

در زمینه‌ی تحلیل داده، محاسبات تعاملی و **تجسم داده‌ها (data visualization)**، پایتون ناگزیر با زبان‌ها و ابزارهای متن‌باز و تجاری دیگری که به‌طور گسترده استفاده می‌شوند—مانند R، MATLAB، SAS، Stata و دیگر ابزارها—مقایسه می‌شود.

در سال‌های اخیر، بهبود چشمگیر کتابخانه‌های متن‌باز پایتون مانند pandas و scikit-learn باعث شده است که پایتون به گزینه‌ای بسیار محبوب برای انجام وظایف تحلیل داده تبدیل شود. ترکیب این توانایی با قدرت کلی پایتون در **مهندسی نرم‌افزار عمومی (general-purpose software engineering)**، آن را به گزینه‌ای عالی برای انتخاب به‌عنوان زبان اصلی در ساخت برنامه‌های داده‌محور بدل کرده است.

## پایتون به‌عنوان «چسب»

بخشی از موفقیت زبان برنامه‌نویسی پایتون در محاسبات علمی، به **سهولت ادغام کدهای C، C++ و FORTRAN** بازمی‌گردد. بسیاری از محیط‌های محاسباتی مدرن مجموعه‌ای مشابه از کتابخانه‌های قدیمی FORTRAN و C را برای انجام محاسباتی مانند جبر خطی، بهینه‌سازی، انتگرال‌گیری، تبدیل فوریه سریع (FFT) و سایر الگوریتم‌های مشابه در اختیار دارند. همین وضعیت در بسیاری از شرکت‌ها و آزمایشگاه‌های ملی نیز دیده می‌شود؛ آن‌ها از پایتون برای «چسباندن» و یکپارچه‌سازی نرم‌افزارهای قدیمی چندین دهه‌ای خود استفاده کرده‌اند.

بسیاری از برنامه‌ها تنها از **بخش‌های کوچکی از کد** تشکیل شده‌اند که بخش عمده‌ی زمان اجرا در آن‌ها صرف می‌شود، در حالی که حجم بزرگی از کد به عنوان **«کد چسب» (glue code)** فقط گاه‌به‌گاه اجرا می‌شود. در بسیاری از موارد، زمان اجرای کد چسب ناچیز است و بهتر است تلاش اصلی روی بهینه‌سازی **گلوگاه‌های محاسباتی** متمرکز شود؛ این بهینه‌سازی گاه با **انتقال بخش‌های حساس به زبان‌های سطح پایین‌تر** مانند C انجام می‌شود.

## حل «مسئله‌ی دو زبان»

در بسیاری از سازمان‌ها معمول است که پژوهش، نمونه‌سازی (Prototype) و آزمون ایده‌های جدید با استفاده از زبان‌های محاسباتی تخصصی‌تری مانند SAS یا R انجام شود و سپس این ایده‌ها برای استفاده در سامانه‌های عملیاتی بزرگ‌تر به زبان‌هایی مانند Java، C# یا C++ منتقل شوند.

با این حال، امروزه بسیاری دریافته‌اند که **پایتون نه‌تنها برای پژوهش و نمونه‌سازی مناسب است، بلکه می‌تواند برای ساخت سامانه‌های عملیاتی نیز به کار رود**. بنابراین، چرا باید دو محیط توسعه‌ی جداگانه نگه داشت وقتی یک محیط کفایت می‌کند؟ باور من این است که شرکت‌های بیشتری به این مسیر روی خواهند آورد، زیرا استفاده از **مجموعه ابزار برنامه‌نویسی یکسان برای پژوهشگران و مهندسان نرم‌افزار** مزایای سازمانی قابل توجهی دارد.

در دهه‌ی گذشته، رویکردهای جدیدی برای حل «مسئله‌ی دو زبان» مطرح شده است، از جمله معرفی Julia (programming language). در بسیاری از موارد، **استفاده‌ی حداکثری از پایتون مستلزم برنامه‌نویسی بخشی از کد در زبان‌های سطح پایین‌تر مانند C یا C++ و سپس ایجاد پیوند (binding) پایتونی برای آن است**. با این حال، فناوری **کامپایلرهای «در لحظه» (Just-in-Time یا JIT)** که توسط کتابخانه‌هایی مانند Numba ارائه می‌شود، این امکان را فراهم کرده است که بسیاری از الگوریتم‌های محاسباتی با کارایی بالا **بدون ترک محیط پایتون** اجرا شوند.

اگرچه پایتون محیطی عالی برای ساخت بسیاری از برنامه‌های تحلیلی و سامانه‌های عمومی است، اما **در برخی کاربردها ممکن است گزینه‌ی ایده‌آلی نباشد**.

</div>

از آنجا که پایتون یک **زبان تفسیری (interpreted)** است، به طور کلی بیشتر کدهای پایتونی **کندتر از کدهای زبان‌های کامپایل‌شده** مانند Java یا C++ اجرا می‌شوند. از آنجا که زمان برنامه‌نویس اغلب ارزشمندتر از زمان CPU است، بسیاری از توسعه‌دهندگان این **مبادله‌ی سرعت در برابر بهره‌وری** را می‌پذیرند. اما در کاربردهایی با **تاخیر بسیار پایین (low latency)** یا **نیاز شدید به بهره‌برداری بهینه از منابع** — مانند یک سامانه‌ی معامله‌گری با فرکانس بالا — صرف زمان برای برنامه‌نویسی در یک زبان سطح پایین‌تر (و البته با بهره‌وری کمتر) مانند C++ به‌منظور دستیابی به بالاترین کارایی ممکن، تصمیمی منطقی است.

یکی دیگر از چالش‌های پایتون، **ساخت برنامه‌های چندنخی (multithreaded) با همزمانی بالا** است، به‌ویژه زمانی که چندین نخ CPU-محور وجود دارد. دلیل این مسئله وجود **قفل مفسر سراسری (Global Interpreter Lock - GIL)** است؛ مکانیزمی که اجازه نمی‌دهد مفسر بیش از یک دستور پایتونی را در یک زمان اجرا کند.

دلایل فنی وجود GIL فراتر از محدوده‌ی این کتاب است. اگرچه در بسیاری از کاربردهای پردازش داده‌های حجیم، از خوشه‌ای از رایانه‌ها برای پردازش داده‌ها در زمان مناسب استفاده می‌شود، اما همچنان مواردی وجود دارد که **یک سامانه‌ی تک‌فرآیندی چندنخی** مطلوب است.

البته این بدان معنا نیست که پایتون نمی‌تواند کد چندنخی واقعی را اجرا کند. **افزونه‌های C پایتون** که از چندریسمانی بومی (در C یا C++) استفاده می‌کنند می‌توانند **کد را به صورت موازی و بدون تأثیر GIL اجرا کنند**، به شرط آنکه نیازی به تعامل مداوم با اشیای پایتونی نداشته باشند.

## NumPy

[NumPy](https://numpy.org/) که مخفف Numerical Python (پایتون عددی) است، برای مدت‌های طولانی یکی از ارکان اصلی محاسبات عددی در پایتون بوده است. این کتابخانه، ساختارهای داده، الگوریتم‌ها و چسب کتابخانه‌ای لازم برای اکثر کاربردهای علمی که شامل داده‌های عددی در پایتون هستند را فراهم می‌کند. NumPy در میان موارد دیگر، حاوی این موارد است:
*   یک شیء آرایه‌ای چندبعدی سریع و کارآمد به نام ndarray
*   توابع برای انجام محاسبات بر اساس عناصر با آرایه‌ها یا عملیات ریاضی بین آرایه‌ها
*   ابزارهایی برای خواندن و نوشتن مجموعه داده‌های مبتنی بر آرایه بر روی دیسک
*   عملیات جبر خطی، تبدیل فوریه و تولید اعداد تصادفی
*   یک API بالغ برای C تا به افزونه‌های پایتون و کدهای C یا C++ بومی امکان دهد تا به ساختارهای داده و امکانات محاسباتی NumPy دسترسی داشته باشند.

فراتر از قابلیت‌های سریع پردازش آرایه که NumPy به پایتون می‌افزاید، یکی از کاربردهای اولیه آن در تحلیل داده، به عنوان یک نگهدارنده (Container) برای داده‌هایی است که قرار است بین الگوریتم‌ها و کتابخانه‌ها منتقل شوند. برای داده‌های عددی، آرایه‌های NumPy برای ذخیره و دستکاری داده‌ها، کارآمدتر از سایر ساختارهای داده داخلی پایتون هستند. همچنین، کتابخانه‌های نوشته شده به زبانهای سطح پایین‌تر مانند C یا FORTRAN می‌توانند روی داده‌های ذخیره شده در یک آرایه NumPy عملیات انجام دهند بدون اینکه نیاز باشد داده را به نمای حافظه دیگری کپی کنند. بنابراین، بسیاری از ابزارهای محاسبات عددی برای پایتون، یا آرایه‌های NumPy را به عنوان یک ساختار داده اولیه فرض می‌گیرند و یا هدفشان قابلیت همکاری با NumPy است.

## pandas

[pandas](https://pandas.pydata.org/) ساختارهای داده و توابع سطح بالایی را ارائه می‌دهد که طراحی شده‌اند تا کار با داده‌های ساختاریافته یا جدولی را شهودی و انعطاف‌پذیر کنند. از زمان ظهورش در سال ۲۰۱۰، به توانمندسازی پایتون برای تبدیل شدن به یک محیط تحلیل داده قدرتمند و پربازده کمک کرده است.
اشیاء اولیه در pandas که در این کتاب استفاده خواهند شد، DataFrame (دیتافریم) که یک ساختار داده جدولی و ستون‌محور با برچسب‌های هر سطر و ستون است، و Series (سری) که یک شیء آرایه‌ای یک‌بعدی دارای برچسب است، می‌باشند.
pandas ایده‌های محاسبات آرایه‌ای NumPy را با قابلیت‌های دستکاری داده موجود در صفحات گسترده (Spreadsheets) و پایگاه‌های داده رابطه‌ای (مانند SQL) در هم می‌آمیزد. این کتابخانه، قابلیت‌های نمایه‌گذاری (Indexing) مناسبی را برای این که بتوانید داده‌ها را تغییر شکل دهید، برش دهید و خرد کنید، تجمع‌ها (Aggregations) را انجام دهید و زیرمجموعه‌هایی از داده را انتخاب کنید، فراهم می‌کند. از آنجایی که دستکاری، آماده‌سازی و پاکسازی داده‌ها از مهارت‌های بسیار مهم در تحلیل داده هستند، pandas یکی از تمرکزهای اصلی این کتاب است.

کمی به پیشینه می‌پردازیم: من ساخت pandas را در اوایل سال ۲۰۰۸ در دوران حضورم در شرکت مدیریت سرمایه AQR، یک شرکت مدیریت سرمایه‌گذاری کمی (Quantitative)، آغاز کردم. در آن زمان، مجموعه‌ای از نیازهای مشخص داشتم که توسط هیچ یک از ابزارهای در دسترسم به خوبی پوشش داده نمی‌شد:

*   ساختارهای داده با محورهای دارای برچسب که از همترازی خودکار یا صریح داده پشتیبانی می‌کنند - این از بروز خطاهای رایج ناشی از داده‌های ناهمتراز و کار با داده‌های با نمایه متفاوت که از منابع مختلف می‌آیند جلوگیری می‌کند.
*   قابلیت‌های یکپارچه سری زمانی
*   ساختارهای داده یکسان که هم داده‌های سری زمانی و هم داده‌های غیر سری زمانی را پردازش می‌کنند.
*   عملیات حسابی و کاهش‌ها (Reductions) که فراداده (Metadata) را حفظ می‌کنند.
*   مدیریت انعطاف‌پذیر داده‌های گمشده (Missing Data)
*   ادغام و سایر عملیات رابطه‌ای موجود در پایگاه‌های داده محبوب (مثلاً مبتنی بر SQL)

من می‌خواستم بتوانم همه این کارها را در یک مکان انجام دهم، ترجیحاً در زبانی که برای توسعه نرم‌افزارهای همه‌منظوره مناسب باشد. پایتون یک زبان نامزد خوب برای این کار بود، اما در آن زمان مجموعه یکپارچه‌ای از ساختارهای داده و ابزار که این قابلیت را فراهم کنند وجود نداشت. در نتیجه، از آنجا که pandas در ابتدا برای حل مسائل مالی و تحلیل کسب‌وکار ساخته شد، دارای قابلیت‌های بسیار عمیق سری زمانی و ابزارهایی است که برای کار با داده‌های دارای نمایه زمانی تولید شده توسط فرآیندهای کسب‌وکار بسیار مناسب هستند.

من بخش بزرگی از سال‌های ۲۰۱۱ و ۲۰۱۲ را صرف گسترش قابلیت‌های pandas به همراه برخی از همکاران سابقم در AQR، آدام کلاین و چنگ شِی، کردم. در سال ۲۰۱۳، من دیگر به اندازه گذشته درگیر توسعه روزمره پروژه شدم و از آن زمان pandas به یک پروژه کاملاً متعلق به جامعه و نگهداری شده توسط جامعه تبدیل شده است که بیش از دو هزار مشارکت‌کننده منحصر به فرد از سراسر جهان دارد.

برای کاربران زبان R برای محاسبات آماری، نام DataFrame آشنا خواهد بود، زیرا این شیء به تقلید از شیء data.frame مشابه در R نام‌گذاری شده است. برخلاف پایتون، data frameها در خود زبان برنامه‌نویسی R و کتابخانه استاندارد آن ساخته شده‌اند. در نتیجه، بسیاری از ویژگی‌های موجود در pandas معمولاً یا بخشی از پیاده‌سازی هسته R هستند یا توسط بسته‌های الحاقی (Add-on Packages) ارائه می‌شوند.

خود نام pandas از اصطلاح panel data (داده‌های تابلویی) که یک اصطلاح اقتصادسنجی برای مجموعه داده‌های ساختاریافته چندبعدی است، گرفته شده و همچنین بازی با عبارت "Python data analysis" (تحلیل داده پایتون) است.

## matplotlib

## IPython و Jupyter

[پروژهٔ **IPython**](https://ipython.org/) در سال ۲۰۰۱ به‌عنوان پروژه‌ای جانبی از سوی «فرناندو پرز» (Fernando Pérez) آغاز شد، با هدف ساخت یک مفسر تعاملی بهتر برای پایتون. در طول بیش از بیست سال بعد، این پروژه به یکی از ابزارهای کلیدی در پشتهٔ دادهٔ مدرن پایتون تبدیل شده است.
اگرچه IPython به‌تنهایی ابزار محاسباتی یا تحلیل داده ارائه نمی‌دهد، اما برای **محاسبات تعاملی** و **توسعهٔ نرم‌افزار** طراحی شده است. این محیط، الگوی کاری **اجرا–کاوش** (execute–explore) را به جای الگوی سنتی **ویرایش–کامپایل–اجرا** (edit–compile–run) در بسیاری از زبان‌های برنامه‌نویسی ترویج می‌کند.
IPython همچنین دسترسی یکپارچه‌ای به **پوستهٔ سیستم‌عامل (Shell)** و **سامانهٔ فایل‌ها (Filesystem)** شما فراهم می‌کند؛ به همین دلیل در بسیاری از موارد نیازی به جابه‌جایی بین پنجرهٔ ترمینال و نشست پایتون نیست.
از آنجا که بخش زیادی از برنامه‌نویسی در تحلیل داده شامل کاوش، آزمون و خطا و تکرار است، IPython می‌تواند باعث **افزایش سرعت انجام کارها** شود.

در سال ۲۰۱۴، فرناندو و تیم IPython پروژهٔ جدیدی به نام [**Jupyter**](https://jupyter.org/) را معرفی کردند؛ ابتکاری گسترده‌تر برای طراحی ابزارهای محاسبات تعاملی که به زبان خاصی وابسته نباشند (**language-agnostic interactive computing tools**).
دفترچهٔ تحت‌وب IPython به **Jupyter Notebook** تبدیل شد که اکنون از بیش از ۴۰ زبان برنامه‌نویسی پشتیبانی می‌کند. امروزه سیستم IPython می‌تواند به‌عنوان یک **کرنل (kernel)** – یعنی حالت زبان برنامه‌نویسی – برای اجرای پایتون در Jupyter استفاده شود.

در حال حاضر IPython خود بخشی از پروژهٔ **متن‌باز Jupyter** است که محیطی پربازده برای محاسبات تعاملی و اکتشافی فراهم می‌کند.
قدیمی‌ترین و ساده‌ترین حالت آن، **پوستهٔ پیشرفتهٔ پایتون (enhanced Python shell)** است که برای تسریع نوشتن، آزمون و اشکال‌زدایی کدهای پایتون طراحی شده است.
شما همچنین می‌توانید از سیستم IPython درون **دفترچهٔ Jupyter** استفاده کنید. سیستم Jupyter Notebook امکان نگارش محتوا با **Markdown** و **HTML** را نیز فراهم می‌کند و بدین ترتیب می‌توانید اسناد غنی شامل **کد و متن** تولید کنید.

من شخصاً در کارهای پایتونی خود به‌طور مرتب از IPython و Jupyter استفاده می‌کنم؛ چه برای اجرای کد، چه اشکال‌زدایی و چه آزمودن برنامه‌ها.

در بخش [منابع همراه کتاب در **GitHub**](https://github.com/wesm/pydata-book)، دفترچه‌های Jupyter حاوی تمام نمونه‌کدهای هر فصل را خواهید یافت.
اگر به GitHub دسترسی ندارید، می‌توانید از [**آینهٔ Gitee** استفاده کنید.](https://gitee.com/wesmckinn/pydata-book)

## SciPy

[**SciPy**](https://scipy.org/) مجموعه‌ای از بسته‌هاست که به طیفی از مسائل بنیادی در محاسبات علمی می‌پردازد. در ادامه برخی از ابزارها و ماژول‌های مهم آن آمده است:

* **scipy.integrate**
  توابع عددی برای **انتگرال‌گیری عددی** و **حل معادلات دیفرانسیل**

* **scipy.linalg**
  روتین‌های **جبر خطی** و **تجزیه‌های ماتریسی** فراتر از آنچه در `numpy.linalg` ارائه می‌شود

* **scipy.optimize**
  الگوریتم‌های **بهینه‌سازی (minimization)** و **یافتن ریشهٔ توابع**

* **scipy.signal**
  ابزارهای **پردازش سیگنال**

* **scipy.sparse**
  **ماتریس‌های تنک (Sparse Matrices)** و حل‌کننده‌های سیستم‌های خطی تنک

* **scipy.special**
  رابطی برای کتابخانهٔ **SPECFUN** نوشته‌شده به زبان FORTRAN، شامل توابع ریاضی متداولی مانند **تابع گاما**

* **scipy.stats**
  توزیع‌های استاندارد پیوسته و گسستهٔ احتمالاتی (توابع چگالی، نمونه‌بردارها، توابع توزیع)، آزمون‌های آماری مختلف و آمار توصیفی

در مجموع، **NumPy** و **SciPy** پایه‌ای محاسباتی کامل و بالغ برای بسیاری از کاربردهای سنتی **محاسبات علمی** فراهم می‌کنند.

## scikit-learn

از زمان آغاز پروژه در سال ۲۰۰۷، [**scikit-learn**](https://scikit-learn.org/) به **اصلی‌ترین ابزار یادگیری ماشین عمومی‌منظور (general-purpose)** برای برنامه‌نویسان پایتون تبدیل شده است.
در زمان نگارش این کتاب، بیش از **دو هزار نفر** در توسعهٔ کد آن مشارکت کرده‌اند.

این کتابخانه شامل زیرماژول‌هایی برای مدل‌های زیر است:

* **دسته‌بندی (Classification):** SVM، نزدیک‌ترین همسایگان (Nearest Neighbors)، جنگل تصادفی (Random Forest)، رگرسیون لجستیک و غیره
* **رگرسیون (Regression):** لاسو (Lasso)، رگرسیون ریج (Ridge Regression) و غیره
* **خوشه‌بندی (Clustering):** k-means، خوشه‌بندی طیفی (Spectral Clustering) و غیره
* **کاهش ابعاد (Dimensionality Reduction):** تحلیل مؤلفه‌های اصلی (PCA)، انتخاب ویژگی، فاکتورگیری ماتریسی و غیره
* **انتخاب مدل (Model Selection):** جستجوی شبکه‌ای (Grid Search)، اعتبارسنجی متقابل (Cross-validation)، معیارهای ارزیابی (Metrics)
* **پیش‌پردازش (Preprocessing):** استخراج ویژگی، نرمال‌سازی و استانداردسازی داده‌ها

در کنار **pandas**، **statsmodels** و **IPython**، کتابخانهٔ scikit-learn نقشی اساسی در تبدیل پایتون به زبانی کارآمد برای **علم داده (Data Science)** داشته است.
اگرچه در این کتاب نمی‌توان به‌طور جامع به تمام امکانات scikit-learn پرداخت، اما مروری کوتاه بر برخی مدل‌ها و نحوهٔ استفاده از آن‌ها همراه با سایر ابزارهای معرفی‌شده ارائه خواهد شد.

## statsmodels

کتابخانهٔ [**statsmodels**](https://statsmodels.org/) برای تحلیل‌های آماری طراحی شده است. این پروژه از کارهای «جاناتان تیلور» (Jonathan Taylor)، استاد آمار دانشگاه استنفورد، آغاز شد؛ او تعدادی از مدل‌های **رگرسیون آماری** محبوب در زبان R را در پایتون پیاده‌سازی کرد.
در سال ۲۰۱۰، «اسکیپر سیبولد» (Skipper Seabold) و «یوزف پرکتولد» (Josef Perktold) به‌صورت رسمی پروژهٔ statsmodels را بنیان‌گذاری کردند و از آن زمان جامعه‌ای فعال از کاربران و مشارکت‌کنندگان حول آن شکل گرفته است.
همچنین «ناتانیل اسمیت» (Nathaniel Smith) پروژهٔ **Patsy** را توسعه داد که چارچوبی برای تعریف فرمول‌ها و مدل‌ها در statsmodels فراهم می‌کند؛ الهام‌گرفته از سیستم فرمول‌نویسی در R.

در مقایسه با scikit-learn، کتابخانهٔ statsmodels بر **آمار کلاسیک (به‌ویژه آماره‌های فراوانی یا Frequentist Statistics)** و **اقتصادسنجی (Econometrics)** تمرکز دارد.
برخی از زیرماژول‌های مهم آن عبارت‌اند از:

* **مدل‌های رگرسیون:** رگرسیون خطی، مدل‌های خطی تعمیم‌یافته (GLM)، مدل‌های خطی مقاوم، مدل‌های خطی آمیخته و غیره
* **تحلیل واریانس (ANOVA)**
* **تحلیل سری‌های زمانی:** مدل‌های AR، ARMA، ARIMA، VAR و دیگر مدل‌ها
* **روش‌های ناپارامتریک:** برآورد چگالی هسته‌ای (Kernel Density Estimation) و رگرسیون هسته‌ای
* **بصری‌سازی نتایج مدل‌های آماری**

statsmodels تمرکز بیشتری بر **استنتاج آماری (Statistical Inference)** دارد و تخمین عدم‌قطعیت و **مقادیر p (p-values)** برای پارامترها ارائه می‌دهد، در حالی‌که scikit-learn بیشتر بر **پیش‌بینی (Prediction)** تمرکز دارد.

همانند scikit-learn، در این کتاب مروری کوتاه بر statsmodels و نحوهٔ استفاده از آن همراه با NumPy و pandas ارائه خواهد شد.

از آنجا که افراد از پایتون برای کاربردهای مختلفی استفاده می‌کنند، هیچ راه‌حل واحدی برای راه‌اندازی پایتون و دریافت بسته‌های الحاقی لازم وجود ندارد. بسیاری از خوانندگان ممکن است یک محیط توسعه کامل پایتون که برای همراهی با این کتاب مناسب باشد را در اختیار نداشته باشند، بنابراین در اینجا دستورالعلی دقیقی برای راه‌اندازی روی هر سیستم عاملی ارائه می‌دهم. من از مینیکندا (Miniconda) که یک نصبِ حداقلی از مدیر بسته کُندا (conda) است، همراه با [کُندا-فورج (conda-forge)](https://conda-forge.org/) که یک توزیع نرم‌افزاری مبتنی بر کُندا و نگهداری‌شده توسط جامعه است، استفاده خواهم کرد. این کتاب در سرتاسر خود از پایتون ۳.۱۰ استفاده می‌کند، اما اگر شما در آینده این متن را می‌خوانید، می‌توانید نسخه جدیدتری از پایتون را نصب کنید.
اگر به هر دلیلی این دستورالعمل‌ها تا زمانی که شما این متن را می‌خوانید منسوخ شده باشند، می‌توانید به [وبسایت من برای این کتاب](https://wesmckinney.com/book) مراجعه کنید که تلاش می‌کنم آن را با آخرین دستورالعمل‌های نصب به‌روز نگه دارم.

## Miniconda روی ویندوز

برای شروع در ویندوز، نصاب مینیکندا را برای آخرین نسخه موجود پایتون (در حال حاضر ۳.۹) از آدرس https://conda.io دانلود کنید. من توصیه می‌کنم دستورالعمل‌های نصب موجود برای ویندوز در وبسایت کُندا را دنبال کنید، زیرا ممکن است از زمان چاپ این کتاب تا زمانی که شما آن را می‌خوانید تغییر کرده باشند.
اکثر افراد نسخه ۶۴ بیتی را می‌خواهند، اما اگر آن روی رایانه ویندوزی شما اجرا نشد، می‌توانید نسخه ۳۲ بیتی را نصب کنید.
هنگامی که از شما پرسیده شد که نصب فقط برای خودتان باشد یا برای تمام کاربران سیستم، گزینه‌ای را انتخاب کنید که برای شما مناسب‌تر است. نصب فقط برای خودتان برای همراهی با کتاب کافی خواهد بود. همچنین از شما پرسیده می‌شود که آیا می‌خواهید مینیکندا را به متغیر محیطی PATH سیستم اضافه کنید یا خیر. اگر این گزینه را انتخاب کنید (من معمولاً این کار را می‌کنم)، ممکن است این نصب مینیکندا، نسخه‌های دیگر پایتونِ نصب‌شده روی سیستم شما را لغو (override) کند. اگر این گزینه را انتخاب نکنید، برای استفاده از این مینیکندا باید از میانبر منوی استارت ویندوز که نصب شده است استفاده کنید. این ورودی در منوی استارت ممکن است "Anaconda3 (64-bit)" نامیده شود.

من فرض می‌کنم که شما مینیکندا را به PATH سیستم خود اضافه نکرده‌اید. برای تأیید صحت پیکربندی، وارد قسمت "Anaconda Prompt (Miniconda3)" در زیر شاخه "Anaconda3 (64-bit)" در منوی استارت شوید. سپس با تایپ کردن `python` سعی کنید مفسر پایتون را راه‌اندازی کنید. باید پیامی مشابه زیر را مشاهده کنید:

```python
(base) C:\Users\Wes>python
Python 3.9 [MSC v.1916 64 bit (AMD64)] :: Anaconda, Inc. on win32
Type "help", "copyright", "credits" or "license" for more information.
>>>
```

## گنو/لینوکس (GNU/Linux)

جزئیات در لینوکس بسته به نوع توزیع لینوکس شما کمی متفاوت خواهد بود، اما در اینجا جزئیاتی برای توزیع‌هایی مانند دبیان، اوبونتو، سنت‌اواس و فدورا ارائه می‌دهم. راه‌اندازی شبیه به مک‌اواس است به جز نحوه نصب مینیکندا. بیشتر خوانندگان خواهان دانلود فایل نصاب پیش‌فرض ۶۴ بیتی خواهند بود که برای معماری x86 است (اما امکان دارد در آینده کاربران بیشتری ماشین‌های لینوکسی مبتنی بر aarch64 داشته باشند). نصاب یک اسکریپت شِل (shell script) است که باید در ترمینال اجرا شود. سپس فایلی با نامی مشابه `Miniconda3-latest-Linux-x86_64.sh` خواهید داشت. برای نصب آن، این اسکریپت را با bash اجرا کنید:

```python
$ bash Miniconda3-latest-Linux-x86_64.sh
```

<div dir="rtl">

برخی از توزیع‌های لینوکس، تمام بسته‌های پایتون مورد نیاز (البته در برخی موارد، نسخه‌های قدیمی) را در مدیران بسته خود دارند و می‌توان آن‌ها را با استفاده از ابزاری مانند `apt` نصب کرد. راه‌اندازی که اینجا توضیح داده شده از مینیکندا استفاده می‌کند، زیرا هم به راحتی در بین توزیع‌های مختلف قابل تکثیر است و هم ارتقای بسته‌ها به آخرین نسخه‌هایشان ساده‌تر است.

شما می‌توانید انتخاب کنید که فایل‌های مینیکندا در کجا قرار گیرند. من توصیه می‌کنم فایل‌ها را در محل پیش‌فرض در پوشه خانگی (home) خود نصب کنید؛ برای مثال، `/home/$USER/miniconda` (طبیعتاً با نام کاربری خودتان).

نصاب از شما خواهد پرسید که آیا مایلید اسکریپت‌های شِل (shell) خود را به گونه‌ای تغییر دهید که مینیکندا به طور خودکار فعال (activate) شود. من به دلایل راحتی، انجام این کار (انتخاب "yes") را توصیه می‌کنم.

پس از اتمام نصب، یک فرآیند ترمینال جدید راه‌اندازی کنید و تأیید کنید که نصب مینیکندای جدید را دریافت می‌کنید:

```python
(base) $ python
Python 3.9 | (main) [GCC 10.3.0] on linux
Type "help", "copyright", "credits" or "license" for more information.
>>>
```

<div dir="rtl">

برای خروج از پوسته پایتون، `exit()` را تایپ کرده و Enter بفشارید یا کلیدهای Ctrl-D را فشار دهید.

## نصب بسته‌های مورد نیاز

اکنون که Miniconda را روی سیستم خود نصب کرده‌ایم، زمان آن رسیده است که بسته‌های اصلی مورد استفاده در این کتاب را نصب کنیم. نخستین گام این است که مخزن `conda-forge` را به عنوان کانال پیش‌فرض بسته‌ها تنظیم کنیم. برای این کار، دستورات زیر را در شِل وارد کنید:


```python
(base) $ conda config --add channels conda-forge
(base) $ conda config --set channel_priority strict
```

<div dir="rtl">

اکنون، یک «محیط» جدید conda با استفاده از دستور `conda create` و نسخه‌ی Python 3.10 ایجاد می‌کنیم:

```python
(base) $ conda create -y -n pydata-book python=3.10
```

<div dir="rtl">

پس از پایان نصب، محیط را با دستور زیر فعال کنید:

```python
(base) $ conda activate pydata-book
(pydata-book) $
```

<div dir="rtl">

ضروری است هر بار که یک ترمینال جدید باز می‌کنید، محیط مورد نظر را با دستور `conda activate` فعال نمایید. می‌توانید هر زمان اطلاعات مربوط به محیط فعال را با دستور زیر مشاهده کنید:

```python
conda info
```

<div dir="rtl">

اکنون بسته‌های اصلی مورد استفاده در سراسر کتاب (به همراه وابستگی‌های آن‌ها) را با دستور زیر نصب می‌کنیم:

```python
(pydata-book) $ conda install -y pandas jupyter matplotlib
```

<div dir="rtl">

در ادامه از برخی بسته‌های دیگر نیز استفاده خواهیم کرد، اما این بسته‌ها را می‌توان بعداً و در زمان نیاز نصب کرد. دو روش برای نصب بسته‌ها وجود دارد: `conda install` و `pip install`. هنگام استفاده از Miniconda، همواره باید `conda install` را در اولویت قرار دهید. اما برخی بسته‌ها از طریق conda در دسترس نیستند؛ در این صورت اگر `conda install $package_name` با خطا مواجه شد، می‌توانید از دستور زیر استفاده کنید:

```python
pip install $package_name
```

<div dir="rtl">

اگر می‌خواهید تمام بسته‌هایی که در ادامه کتاب مورد استفاده قرار می‌گیرند را یک‌جا نصب کنید، می‌توانید دستور زیر را اجرا کنید:

```python
conda install lxml beautifulsoup4 html5lib openpyxl \
    requests sqlalchemy seaborn scipy statsmodels \
    patsy scikit-learn pyarrow pytables numba
```

<div dir="rtl">

در سیستم‌عامل ویندوز، به جای کاراکتر `\` (که برای ادامه خط در لینوکس و macOS استفاده می‌شود) از علامت `^` استفاده کنید.

برای به‌روزرسانی بسته‌ها از دستور زیر استفاده می‌شود:

```python
conda update package_name
```

<div dir="rtl">

همچنین `pip` نیز قابلیت ارتقای بسته‌ها را با استفاده از گزینه `--upgrade` دارد:

```python
pip install --upgrade package_name
```

<div dir="rtl">

در طول کتاب فرصت‌های متعددی برای تمرین این دستورات خواهید داشت.

> ⚠️ توجه:
> گرچه می‌توانید از هر دو ابزار conda و pip برای نصب بسته‌ها استفاده کنید، توصیه می‌شود از به‌روزرسانی بسته‌هایی که با conda نصب شده‌اند با pip (و برعکس) خودداری کنید، زیرا این کار می‌تواند باعث بروز مشکلاتی در محیط شما شود. بهترین روش این است که تا حد امکان از conda استفاده کنید و تنها در صورتی که بسته‌ای با conda در دسترس نبود، از pip بهره ببرید.

# 1.5 جامعه و کنفرانس‌ها

فراتر از یک جست‌وجوی اینترنتی ساده، فهرست‌های پستی (mailing lists) مرتبط با پایتون در حوزهٔ علم و داده معمولاً بسیار مفید و پاسخ‌گو به پرسش‌ها هستند. برخی از فهرست‌های مهم که ارزش بررسی دارند عبارت‌اند از:

* **pydata**: یک گروه گوگل برای پرسش‌های مربوط به پایتون در تحلیل داده و کتابخانهٔ pandas
* **pystatsmodels**: برای پرسش‌های مرتبط با statsmodels یا pandas
* فهرست پستی scikit-learn ([scikit-learn@python.org](mailto:scikit-learn@python.org)) برای پرسش‌های مرتبط با یادگیری ماشین در پایتون
* **numpy-discussion**: برای پرسش‌های مربوط به NumPy
* **scipy-user**: برای پرسش‌های عمومی دربارهٔ SciPy یا مباحث علمی مرتبط با پایتون

من عمداً نشانی‌های اینترنتی (URL) این فهرست‌ها را ذکر نکردم، زیرا ممکن است در آینده تغییر کنند. این فهرست‌ها را به‌راحتی می‌توان از طریق یک جست‌وجوی اینترنتی یافت.

هر ساله کنفرانس‌های متعددی در سراسر جهان برای برنامه‌نویسان پایتون برگزار می‌شود.
اگر تمایل دارید با برنامه‌نویسان پایتونی که علایق مشابهی با شما دارند ارتباط برقرار کنید، توصیه می‌کنم در صورت امکان، شرکت در یکی از این کنفرانس‌ها را در نظر بگیرید. بسیاری از این رویدادها دارای برنامه‌های حمایتی مالی هستند تا افرادی که توانایی پرداخت هزینهٔ ثبت‌نام یا سفر را ندارند نیز بتوانند شرکت کنند. برخی از کنفرانس‌های مهم عبارت‌اند از:

* **PyCon** و **EuroPython**: دو کنفرانس اصلی و عمومی پایتون به‌ترتیب در آمریکای شمالی و اروپا
* **SciPy** و **EuroSciPy**: کنفرانس‌های متمرکز بر محاسبات علمی در آمریکای شمالی و اروپا
* **PyData**: مجموعه‌ای جهانی از کنفرانس‌های منطقه‌ای با تمرکز بر علم داده و کاربردهای تحلیل داده
* کنفرانس‌های بین‌المللی و منطقه‌ای PyCon (برای مشاهدهٔ فهرست کامل به [https://pycon.org](https://pycon.org) مراجعه کنید)
